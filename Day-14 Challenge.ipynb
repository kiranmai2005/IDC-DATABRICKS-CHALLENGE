```python
from transformers import pipeline
import mlflow

# Load pre-trained sentiment model
classifier = pipeline("sentiment-analysis")

reviews = [
    "This product is amazing!",
    "Terrible quality, waste of money",
    "Decent value for the price",
    "I will never buy this again"
]

results = classifier(reviews)

results
```

```python
with mlflow.start_run(run_name="sentiment_analysis_demo"):

    mlflow.log_param("model", "distilbert-base-uncased-finetuned-sst-2-english")
    mlflow.log_param("task", "sentiment-analysis")

    positive = sum(1 for r in results if r["label"] == "POSITIVE")
    negative = sum(1 for r in results if r["label"] == "NEGATIVE")

    mlflow.log_metric("positive_reviews", positive)
    mlflow.log_metric("negative_reviews", negative)

    mlflow.log_text(str(results), "sample_predictions.txt")
```
