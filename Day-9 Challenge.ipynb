```python
%sql
USE CATALOG ecommerce;
USE SCHEMA silver;
```

```python
%sql
WITH daily_revenue AS (
  SELECT
    event_date,
    SUM(price) AS revenue
  FROM silver.events
  WHERE event_type = 'purchase'
  GROUP BY event_date
)
SELECT
  event_date,
  revenue,
  AVG(revenue) OVER (
    ORDER BY event_date
    ROWS BETWEEN 6 PRECEDING AND CURRENT ROW
  ) AS revenue_7d_ma
FROM daily_revenue
ORDER BY event_date;
```

```python
%sql
SELECT
  price_tier,
  SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END) AS views,
  SUM(CASE WHEN event_type = 'purchase' THEN 1 ELSE 0 END) AS purchases,
  ROUND(
    SUM(CASE WHEN event_type = 'purchase' THEN 1 ELSE 0 END) * 100.0 /
    NULLIF(SUM(CASE WHEN event_type = 'view' THEN 1 ELSE 0 END), 0),
    2
  ) AS conversion_rate
FROM ecommerce.silver.events
GROUP BY price_tier
ORDER BY conversion_rate DESC;
```

```python
%sql
SELECT
  price_tier,
  COUNT(CASE WHEN event_type = 'view' THEN 1 END) AS views,
  COUNT(CASE WHEN event_type = 'purchase' THEN 1 END) AS purchases,
  ROUND(
    COUNT(CASE WHEN event_type = 'purchase' THEN 1 END) * 100.0 /
    NULLIF(COUNT(CASE WHEN event_type = 'view' THEN 1 END), 0),
    2
  ) AS conversion_rate
FROM ecommerce.silver.events
GROUP BY price_tier
ORDER BY conversion_rate DESC;
```

```python
%sql
SELECT
  product_id,
  SUM(price) AS revenue,
  COUNT(*) AS purchases
FROM ecommerce.silver.events
WHERE event_type = 'purchase'
GROUP BY product_id
ORDER BY revenue DESC
LIMIT 10;
```
