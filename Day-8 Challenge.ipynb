```python
%sql
-- Create catalog
CREATE CATALOG IF NOT EXISTS ecommerce;

USE CATALOG ecommerce;

-- Create schemas
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
```

```python
%sql
-- Bronze table
CREATE TABLE IF NOT EXISTS bronze.events (
    user_id STRING,
    product_id STRING,
    event_type STRING,
    event_time TIMESTAMP,
    price DOUBLE,
    user_session STRING,
    ingestion_ts TIMESTAMP
) USING DELTA;

-- Silver table
CREATE TABLE IF NOT EXISTS silver.events (
    user_id STRING,
    product_id STRING,
    event_type STRING,
    event_time TIMESTAMP,
    price DOUBLE,
    user_session STRING,
    ingestion_ts TIMESTAMP,
    event_date DATE,
    price_tier STRING
) USING DELTA;

-- Gold table
CREATE TABLE IF NOT EXISTS gold.products (
    product_id STRING,
    product_name STRING,
    views BIGINT,
    purchases BIGINT,
    revenue DOUBLE,
    conversion_rate DOUBLE
) USING DELTA;
```
